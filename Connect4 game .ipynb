```python
# This is a basic implementation of the Connect-4 game in Python

# define the board as a list of lists
board = [[' ' for x in range(7)] for y in range(6)]

# define the players
players = ['X', 'O']

# define the function to display the board
def display_board():
    for row in board:
        print(row)

# define the function to check for a win
def check_win():
    # check for horizontal win
    for row in range(len(board)):
        for col in range(len(board[row]) - 3):
            if board[row][col] == board[row][col + 1] == board[row][col + 2] == board[row][col + 3] and board[row][col] != ' ':
                return True

    # check for vertical win
    for row in range(len(board) - 3):
        for col in range(len(board[row])):
            if board[row][col] == board[row + 1][col] == board[row + 2][col] == board[row + 3][col] and board[row][col] != ' ':
                return True

    # check for diagonal win
    for row in range(len(board) - 3):
        for col in range(len(board[row]) - 3):
            if board[row][col] == board[row + 1][col + 1] == board[row + 2][col + 2] == board[row + 3][col + 3] and board[row][col] != ' ':
                return True
            if board[row][col + 3] == board[row + 1][col + 2] == board[row + 2][col + 1] == board[row + 3][col] and board[row][col + 3] != ' ':
                return True

    return False

# define the function to check if the board is full
def check_full():
    for row in range(len(board)):
        for col in range(len(board[row])):
            if board[row][col] == ' ':
                return False
    return True

# define the main game loop
def play_game():
    current_player = 0
    while True:
        display_board()
        print(f"Player {players[current_player]}, select a column to drop your piece (0-6)")
        column = int(input())
        for row in range(len(board) - 1, -1, -1):
            if board[row][column] == ' ':
                board[row][column] = players[current_player]
                break
        if check_win():
            display_board()
            print(f"Player {players[current_player]} wins!")
            break
        if check_full():
            display_board()
            print("It's a tie!")
            break
        current_player = (current_player + 1) % 2

# start the game
play_game()
```
